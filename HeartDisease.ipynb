```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(df3)
print(f"Explained Variance Ratio: {pca.explained_variance_ratio_}")

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1],c=df3['target'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Heat Disease dataset')
plt.colorbar()
plt.show()
```

```python
pca_df = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])

pca_df.head
```

```python
print("Available columns in the dataset:")
print(pca_df.columns)
pca_df.head()
```

```python
import numpy as np
from sklearn.cluster import KMeans

k_range = range(1, 8)
wcss = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(pca_df)
    wcss.append(kmeans.inertia_)
    print(f'the optimal number of K {k}, WCSS: {kmeans.inertia_}')

# Elbow Method visualization
plt.figure(figsize=(8, 5))
plt.plot(k_range, wcss, marker='o', linestyle='--', color='b')
plt.title('Elbow Method to Determine Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.xticks(k_range)
plt.grid()
plt.show()
```

```python
from sklearn.ensemble import IsolationForest

# identify outliers
iso_forest = IsolationForest(contamination=0.1, random_state=42)
df['anomaly_isoforest'] = iso_forest.fit_predict(pca_df)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Assuming `True_Label` is available in the DataFrame
# Apply K-means clustering
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(pca_df[['PCA1', 'PCA2']])

# Add cluster labels to the DataFrame
pca_df['Cluster'] = clusters

# Map cluster labels to readable categories
cluster_labels = {0: "No Heart Disease", 1: "Have Heart Disease"}
pca_df['Cluster_Label'] = pca_df['Cluster'].map(cluster_labels)

# Align cluster labels with true labels using the Hungarian method
from scipy.optimize import linear_sum_assignment

def match_labels(true_labels, predicted_labels):
    # Create a confusion matrix
    contingency_matrix = confusion_matrix(true_labels, predicted_labels)
    # Solve the assignment problem
    row_ind, col_ind = linear_sum_assignment(-contingency_matrix)
    # Create a mapping from predicted labels to true labels
    mapping = {col: row for row, col in zip(row_ind, col_ind)}
    # Apply the mapping to the predicted labels
    return np.array([mapping[label] for label in predicted_labels])

# Match the clusters to true labels
adjusted_clusters = match_labels(df3['target'], pca_df['Cluster'])

# Update the DataFrame with adjusted clusters
pca_df['Adjusted_Cluster'] = adjusted_clusters

# Calculate metrics
accuracy = accuracy_score(df3['target'], pca_df['Adjusted_Cluster'])
precision = precision_score(df3['target'], pca_df['Adjusted_Cluster'])
recall = recall_score(df3['target'], pca_df['Adjusted_Cluster'])
f1 = f1_score(df3['target'], pca_df['Adjusted_Cluster'])

# Print metrics
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")

# Print classification report
print("\nClassification Report:")
print(classification_report(df3['target'], pca_df['Adjusted_Cluster']))

# Add cluster centers to the plot
centers = kmeans.cluster_centers_

# Visualize clusters on a 2D PCA plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='PCA1',
    y='PCA2',
    hue='Cluster_Label',
    palette='Set1',
    data=pca_df,
    legend='full'
)
plt.scatter(centers[:, 0], centers[:, 1], s=100, c='g', marker='X', label='Centroids')
plt.title('2D PCA Clustering', fontsize=16)
plt.xlabel('PCA1', fontsize=14)
plt.ylabel('PCA2', fontsize=14)
plt.grid(alpha=0.5)
plt.legend(title='Cluster', fontsize=10, title_fontsize=12)
plt.show()

# Analyze cluster characteristics
cluster_0 = pca_df[pca_df['Adjusted_Cluster'] == 0]
cluster_1 = pca_df[pca_df['Adjusted_Cluster'] == 1]

print("Cluster 0 Characteristics:")
print(cluster_0.describe())

print("\nCluster 1 Characteristics:")
print(cluster_1.describe())
```
